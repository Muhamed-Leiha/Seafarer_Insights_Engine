```python
import pandas as pd
import numpy as np


df = pd.read_csv('Titanic.csv')

# Task: Splitting the dataset and re-combining it using concatenation 
df1 = df.iloc[:500]
df2 = df.iloc[500:]
df_recreated = pd.concat([df1, df2], axis=0)

# Task: Generating random nationalities and merging them with passenger detials
np.random.seed(42)
nationalities = ['British', 'American', 'French', 'German', 'Irish', 'Swedish']
passenger_ids = df['PassengerId'].sample(frac=0.7, random_state=42).tolist()
nationality_list = np.random.choice(nationalities, size=len(passenger_ids))

passenger_details = pd.DataFrame({
    'PassengerId': passenger_ids,
    'Nationality': nationality_list
})

df_with_nationality = pd.merge(df, passenger_details, on='PassengerId', how='left')

# Task: Group by Pclass and Sex
grouped_stats = df.groupby(['Pclass', 'Sex']).agg({
    'Age': 'mean',
    'Fare': 'sum'
}).reset_index()

# Task: Pivot table for survival rate
survival_pivot = pd.pivot_table( df , values='Survived' , index='Sex' , columns='Pclass' , aggfunc='mean')

# Task: Add synthetic date and compute monthly survival
df['Date'] = pd.date_range(start='1912-04-10', periods=len(df), freq='D')
df['Date'] = pd.to_datetime(df['Date'])
df['YearMonth'] = df['Date'].dt.to_period('M')

monthly_survival = df.groupby('YearMonth').agg({
    'Survived': ['sum', 'count', 'mean']
}).reset_index()


monthly_survival.columns = ['YearMonth', 'Survival_Count', 'Total_Passengers', 'Survival_Rate']
# Task: Displeying aggregated statistics and multi-level data summarise (Pivot Table)

print("Q1 - First few rows of recreated DataFrame:")
print(df_recreated.head())
print("\n" + "="*50)

print("Q2 - DataFrame with Nationality (first few rows):")
print(df_with_nationality[['PassengerId', 'Name', 'Nationality']].head())
print("\n" + "="*50)

print("Q3 - Grouped Statistics:")
print(grouped_stats)
print("\n" + "="*50)

print("Q4 - Survival Rate Pivot Table:")
print(survival_pivot)
print("\n" + "="*50)

print("Q5 - Monthly Survival Statistics:")
print(monthly_survival)
```
